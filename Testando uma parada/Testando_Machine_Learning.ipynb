# Learning ML

```python
import pandas as pd
arquivo = pd.read_csv('wine_dataset.csv')
```

```python
arquivo
```

```python
arquivo['style'] = arquivo['style'].replace('red',0) #o Valor red na coluna style recebe 0
```

```python
arquivo['style'] = arquivo['style'].replace('white',1) # o Valor white na coluna style recebe 1
```

```python
arquivo['style']
```

```python
#Separando as variÃ¡veis entre preditoras e variÃ¡vel alvo
y = arquivo['style']
x = arquivo.drop('style', axis = 1) # x recebe tudo menos a coluna style, (axis = 1 Ã© coluna)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split #Importando funÃ§Ã£o do sklearn chamada train_test_split

#criando conjunto de dados de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3) #Quero que 30% dos dados sejam para teste
```

```python
from sklearn.ensemble import ExtraTreesClassifier
#CriaÃ§Ã£o de modelo
modelo = ExtraTreesClassifier()
modelo.fit(x_treino,y_treino) #fit Ã© treinar
 
#Imprimindo resultado
resultado = modelo.score(x_teste,y_teste)
print('AcurÃ¡cia: ', resultado)
```
